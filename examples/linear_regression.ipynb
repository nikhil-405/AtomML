The usage for linear regression is identical to the sklearn API

```python
import numpy as np
from utils.data_loaders import load_tabular_data ## for loading data
from regression import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
# Data Loading --> you just need to pass in x, y as Pandas Series objects or np arrays
file_path = r"../data/simple_linear_regression_data.csv"
x, y = load_tabular_data(file_path, x = ["x1", "x2", "x3"], y = ["x4"])
print(type(x), type(y))
x = StandardScaler().fit_transform(x)
```

```python
# create an instance of the class
model = LinearRegression(lr = 0.1, n_iters = 1000)

# fit method
model.fit(x, y)

# predict method
predictions = model.predict(x)
```

```python
import matplotlib.pyplot as plt

y_true = y.flatten()
y_pred = predictions.flatten()

plt.scatter(y_true, y_pred)
mn, mx = min(y_true.min(), y_pred.min()), max(y_true.max(), y_pred.max())
plt.plot([mn, mx], [mn, mx])
plt.xlabel("True target")
plt.ylabel("Predicted target")
plt.title("True vs. Predicted")
plt.grid(True)
plt.show()
```

```python
residuals = y_true - y_pred

plt.hist(residuals, bins=20)
plt.xlabel("Residual (true − pred)")
plt.ylabel("Count")
plt.title("Residuals Distribution")
plt.grid(True)
plt.show()
```

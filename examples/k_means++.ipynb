```python
import numpy as np
from clustering import KMeansPlusPlus
import matplotlib.pyplot as plt
```

```python
X, _ = __import__('sklearn.datasets', fromlist = ['make_blobs']).make_blobs(
        n_samples = 300, centers = 4, cluster_std = 0.60, random_state = 42
    )
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.title("Sample Data")
plt.show()
```

```python
kmeans = KMeansPlusPlus(X)
centers,clusters = kmeans.cluster(k = 4, max_iters = 5)
```

```python
# Plot the clusters
plt.scatter(X[:, 0], X[:, 1])
centers = np.array(centers)
plt.scatter(centers[:, 0], centers[:, 1], marker = 'X')
plt.title("K-Means++ Clustering")
```
